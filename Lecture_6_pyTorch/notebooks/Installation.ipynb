# Installation

## Create a conda environment and activate it (in an anaconda command shell)

    conda create -n pytorch python=3.11
    conda activate pytorch



Go to the following web-page:

    https://pytorch.org/get-started/locally/

Select the good configuration of your platform.

if Linux or Windows and you want GPU support (offered through NVIDIA)

    https://developer.nvidia.com/cuda-downloads

--> select Cuda version, but also install cuda support on your machine.

GPU (CUDA) support is only available on Linux and PC platforms.

Download and install by copy pasting the command to run. An example for a Windows machine, using only CPU:

    conda instal pytorch torchvision torchaudio cpuonly -c pytorch
    
Or using with GPU support, but be sure to install CUDA support **first**!

    conda install pytorch torchvision torchaudio cudatoolkit=10.1 -c pytorch

You can also upload in a Google Colab
